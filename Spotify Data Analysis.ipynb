Correlation Heatmap: It visualizes the relationship between different variables like popularity, danceability, energy, etc., with color coding to indicate the strength and direction of the correlations.

Color Coding: Dark redand yellow  shows strong positive correlations, and dark blu[purple]  indicates strong negative correlations.
Variables Analyzed: The heatmap includes various attributes of songs, such as popularity, danceability, energy, loudness, and other
s.
Outcome Interpretation: The numerical values within the heatmap cells represent correlation coefficients, which quantify the degree to which two variables are relat
ed.
The outcome of this analysis helps to understand how different characteristics of songs relate to each other, which can be valuable for tasks like music recommendation or market analysis. For instance, a strong positive correlation between popularity and danceability might suggest that more danceable songs tend to be more popular.

```python
numeric_columns = df_tracks.select_dtypes(include=['number']).columns
corr_df=df_tracks[numeric_columns].drop(["key","mode","explicit"],axis=1).corr(method="pearson")
plt.figure(figsize=(14,6))
heatmap=sns.heatmap(corr_df,annot=True,fmt=".1g", vmin=-1, vmax=1, center=0, cmap="inferno", linewidths =1, linecolor="black")
heatmap.set_title("correlation Heatmap Between variable")
heatmap.set_xticklabels(heatmap.get_xticklabels(), rotation=90)
```

```python
corr_df.head()
```

Popularity and Loudness have a positive correlation of approximately 0.327, suggesting that louder songs tend to be more popular.

Danceability and Valence (the musical positiveness conveyed by a song) have a strong positive correlation of 0.528, indicating that songs that are more danceable tend to be more positive.

Energy and Acousticness have a strong negative correlation of approximately -0.715, meaning that songs with higher energy tend to be less acoustic.

These correlations can be useful for understanding the relationships between different characteristics of songs and can inform decisions in areas such as marketing strategies or playlist curation.

`

- `plt.figure(figsize=(10,6))`: This line of code sets the size of the figure that will be created. The `figsize` parameter takes a tuple that defines the width and height in inches.
- `sns.regplot()`: This function creates a regression plot. It requires the dataset (`sample_df`) and the two variables you want to analyze (`x` for the independent variable and `y` for the dependent variable). In this case, it's plotting the relationship between 'energy' and 'loudness' of songs.
- `color="c"`: This sets the color of the plot to cyan.
- `.set(title = "Loudness vs Energy correlation")`: This method sets the title of the plot.

The regression plot will show the scatter of points representing the 'energy' and 'loudness' data points, along with a line of best fit that indicates the trend or relationship between these two variables. The plot helps in understanding how the loudness of a song is related to its energy level; a positive slope would suggest that as energy increases, loudness tends to increase as well.

```python
plt.figure(figsize=(10,6))
sns.regplot(data = sample_df , y = "loudness", x = "energy", color = "c").set(title = "Loudness vs Energy correlation")
```
